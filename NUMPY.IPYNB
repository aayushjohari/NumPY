## MATHEMATICAL OPERATIONS

ADDITION

```python
np.random.seed(3)

arr = np.random.randint(1,10,(3,5))
print(arr)
print(arr +1)
```

```python
np.random.seed(3)

arr1 = np.random.randint(1,10,(3,5))

arr2 = np.random.randint(1,10,(3,5))
print(arr1)
print("-"*50)
print(arr2)
print("-"*50)
print(arr1+ arr2)
```

subtraction

```python
np.random.seed(3)

arr1 = np.random.randint(1,10,(3,5))

print(arr1)
print("-"*50)
print(arr1-1)
```

##  MATRIX MULTPLICATION


```python
np.random.seed(3)

arr1 = np.random.randint(1,10,(2,2))
arr2 = np.random.randint(1,10,(2,2))

print(arr1)
print("-"*50)
print(arr2)
print("-"*50)
print(arr1@arr2)
```

## MEAN , vARINACE AND STANDARD DEVIATION


```python
np.random.seed(3)
arr = np.random.randint(1,10,10)
print(arr)

print(arr.mean())
print(arr.var())
print(arr.std())
```

## FILTERING

```python
np.random.seed(3)
arr = np.random.randint(1,10,10)
print(arr)

print(arr[arr>5])
```

##  MERGING ARRAYS
CANCATENATE

```python
np.random.seed(3)
arr1 = np.random.randint(1,10,(2,4))
arr2 = np.random.randint(1,10,(2,4))

print(np.concatenate((arr1, arr2)))
```
